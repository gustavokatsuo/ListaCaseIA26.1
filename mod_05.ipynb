### **Redução de Dimensionalidade**

**3.** O que é a "Maldição da Dimensionalidade"? Explique como o aumento excessivo de variáveis (features) afeta algoritmos baseados em distância.

**4.** Contraste as abordagens de **Projeção** e **Manifold Learning**. Qual é a premissa principal de cada uma ao tentar reduzir a dimensionalidade dos dados?

**5.** No algoritmo **PCA (Principal Component Analysis)**, o que representam os "componentes principais"? Descreva brevemente o que o primeiro e o segundo componentes buscam maximizar.

**6.** Descreva os 5 passos matemáticos fundamentais para a execução do algoritmo PCA (desde a centralização dos dados até a projeção final).

**7.** O que é um *Scree Plot* e como ele, em conjunto com a métrica de Variância Acumulada, ajuda o cientista de dados a decidir o número ideal de componentes $k$ a serem mantidos após a aplicação do PCA?

**8.** Explique a intuição por trás do algoritmo **LLE (Locally Linear Embedding)**. Por que ele é mais indicado do que o PCA para datasets com estruturas complexas, como o clássico *Swiss Roll*?

**9.** Faça um paralelo rápido entre o **t-SNE** e os **Autoencoders**. Qual a principal finalidade de uso do t-SNE e por que não o utilizamos diretamente para transformar novos dados de teste em um pipeline de produção?

### **Clustering**

**10.** O conceito de "similaridade" é central em problemas de clustering. Cite e explique brevemente as quatro noções de similaridade mencionadas no material (proximidade euclidiana, densidade, conectividade e probabilidade).

**11.** Descreva o funcionamento iterativo do algoritmo **K-Means** (os passos de *Atribuição* e *Atualização de Centroides*). Quais são as suas duas principais limitações quanto à forma geométrica dos clusters e à configuração inicial?

**12.** Como funciona o *Elbow Method* (Método do Cotovelo)? Explique como ele utiliza a "inércia" para ajudar a determinar o número ideal de clusters no K-Means.

**13.** Além de observar visualmente os agrupamentos, precisamos de métricas matemáticas para avaliar a qualidade dos clusters. Explique o que o **Silhouette Score** mede e como interpretar seus valores (de -1 a 1).

**14.** O algoritmo **DBSCAN** baseia-se em densidade. Defina o que são os hiperparâmetros `eps` e `min_samples` e explique como o algoritmo identifica e classifica pontos como *outliers*.

**15.** Como a heurística do gráfico *K-Distance* (K-Distance Graph) é utilizada para auxiliar na escolha do hiperparâmetro `eps` perfeito para o DBSCAN? O que devemos buscar no gráfico?

**16.** Qual é a vantagem da abordagem probabilística do **GMM (Gaussian Mixture Models)** em relação à atribuição rígida ("hard assignment") feita pelo K-Means?

**17.** No GMM, não utilizamos o *Elbow Method*. Quais critérios estatísticos utilizamos para penalizar a complexidade do modelo e escolher o número ótimo de componentes? Explique o que se busca nos valores dessas métricas.
